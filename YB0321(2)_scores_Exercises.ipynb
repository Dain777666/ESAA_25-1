````python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}

df = pd.DataFrame(raw_data)
print(df)
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
# 산점도 그리기
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['preTestScore'], df['postTestScore'],
                      s=df['age']*5,  # 나이에 따라 점 크기 조정
                      alpha=0.6)

# 축 라벨 및 제목
plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.title('Pre-Test vs Post-Test Scores (Size ~ Age)')

# 각 점에 이름(label) 추가
for i in range(len(df)):
    plt.text(df['preTestScore'][i] + 0.5,
             df['postTestScore'][i] + 0.5,
             df['first_name'][i], fontsize=9)

plt.grid(True)
plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
# 산점도
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['preTestScore'],
                      df['postTestScore'],
                      s=df['postTestScore'] * 4.5,    # 점 크기
                      c=df['female'],                 # 색상 (0=남, 1=여)
                      cmap='coolwarm',                # 색상 맵 (여성-빨강, 남성-파랑)
                      alpha=0.6)

# 축 라벨과 제목
plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.title('Pre-Test vs Post-Test Scores / (Size ~ Post-Test Score, Color ~ Gender)')

plt.legend()

plt.grid(True)
plt.show()
````

### BONUS: Create your own question and answer it.

> "누가 사전(preTestScore) 대비 사후(postTestScore) 점수를 가장 많이 향상시켰을까?"

(즉, postTestScore - preTestScore가 가장 큰 사람은 누구인가?)

````python
# 향상도 계산
df['improvement'] = df['postTestScore'] - df['preTestScore']

# 향상도 최대인 사람 찾기
max_improve = df.loc[df['improvement'].idxmax()]

print("가장 많이 향상된 사람은:", max_improve['first_name'])
print("향상도:", max_improve['improvement'])
````
