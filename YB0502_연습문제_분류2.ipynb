# 2. XGBoost

```python
import requests
import zipfile
import os

# 1. 다운로드할 URL 지정
url = 'https://bit.ly/3i4n1QB'
zip_path = 'data.zip'

# 2. 파일 다운로드
response = requests.get(url)
with open(zip_path, 'wb') as f:
    f.write(response.content)

# 3. 압축 해제
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('data')  # 'data' 폴더에 압축 해제

# (선택) 압축 파일 삭제
os.remove(zip_path)
```

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
from xgboost import XGBClassifier

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

train.drop('index',axis = 1 ,inplace =True)
test.drop('index',axis = 1 ,inplace =True)
```

```python
# 데이터 확인
train.info()
```

```python
train.head()
```

```python
train.type.value_counts()
```

```python
# 원핫 인코딩 (pd.get_dummies())

train_encoded = pd.get_dummies(train, columns=['type'])
test_encoded = pd.get_dummies(test, columns=['type'])

# 확인
train_encoded.head()
```

```python
# 모델 정의
import pandas as pd
from xgboost import XGBClassifier
```

```python
# 모델 학습
# X 는 train에서 quality 를 제외한 모든 변수
# y 는 train의 qulity 변수

train_x = train_encoded.drop(['quality'], axis=1)
train_y = train_encoded['quality']
test_x = test_encoded

from sklearn.preprocessing import LabelEncoder

# LabelEncoder로 클래스 인코딩 (3~9 → 0~6)
le = LabelEncoder()
train_y_encoded = le.fit_transform(train_y)


# fit 메소드를 이용해 모델 학습
xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')
xgb.fit(train_x, train_y_encoded)
```

```python
# predict 메소드와 원핫 인코딩한 test 데이터를 이용해 품질 예측

xgb_preds = xgb.predict(test_x)
```

```python
# sample_submission.csv 파일을 불러와 예측된 값으로 채워 주기

smpl_sub = pd.read_csv('sample_submission.csv')
smpl_sub['quality'] = xgb_preds
```
