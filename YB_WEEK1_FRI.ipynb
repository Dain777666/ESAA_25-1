✋DataFrame iloc[ ] 연산자



*   iloc[ ]는 위치(Location) 기반 인덱싱 방식으로 동작
>  위치 기반 인덱싱은 행과 열 위치를, 0을 출발점으로 하는 세로축, 가로축 좌표 정숫값으로 지정하는 방식
*   loc[ ]는 명칭(Label)기반 인덱싱 방식으로 동작
> 명칭 기반 인덱싱은 데이터 프레임의 인덱스 값으로 행 위치를, 칼럼의 명칭으로 열 위치를 지정하는 방식

✋**결손 데이터 처리하기**



*   결손 데이터:  칼럼에 값이 없는, 즉 NULL인 경우를 의미하며, 이를 넘파이의 NaN으로 표시한다.
*   기본적으로 머신러닝 알고리즘은 이 NaN 값을 처리하지 않으므로 이 값을 다른 값으로 대체해야한다.
*   NaN 여부를 확인하는 API는 isna( )이며, NaN 값을 다른 값으로 대체하는 API는 fillna()이다.






````python
##isna()로 결손 데이터여부 확인##

titanic_df.isna().head(3)
````

````python
titanic_df.isna().sum()

#sum()을 호출 시 True는 내부적으로 숫자 1로, False는 숫자 0으로 변환
````

````python
##fillna()로 결손 데이터 대체하기##

#‘Cabin’ 칼럼의 NaN 값을‘C000’으로 대체
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('c000')
titanic_df.head(3)
````

````python
#주의점: fillna()를 이용해 반환 값을 다시 받거나 inplace=True 파라미터를 fillna()에 추가해야 실제 데이터 세트 값이 변경된다.
#ex) titanic_df[‘Cabinl].fillna(,COOO,, inplace=True)
````

````python
#‘Age’칼럼의 NaN 값을 평균 나이로,‘Embarked’칼럼의 NaN 값을 ‘으로 대체

titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
titanic_df.isna().sum()
````

````python
##apply lambda 식으로 데이터 가공##

###lambda###
def get_square(a):
  return a**2

print('3의 제곱은:', get_square(3))
````

````python
#lambda: 함수의 선언과 함수 내의 처리를 한 줄의 식으로 쉽게 변환하는 식

lambda_square = lambda x : x ** 2
print('3의 제곱은:', lambda_square(3))
````

````python
#여러 개의 값을 입력 인자로 사용해야 할 경우에는 보통 map() 함수를 결합해서 사용

a = [1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
````

````python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head(3)
````

````python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
````

````python
#if 절의 경우 if 식보다 반환 값을 먼저 기술해야 함
#if, else만 지원하고 if, else if, else와 같이 else if는 지원하지 않음
# else if를 이용하기 위해서는 else 절을 ()로 내포해 () 내에서 다시 if else 적용해야 함

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
````

````python
#나이에 따라 세분화된 분류를 수행하는 함수 생성

def get_category(age):
  cat = ''
  if age <= 5 : cat = 'Baby'
  elif age <= 12 : cat = 'Child'
  elif age <= 18: cat = 'Teenager'
  elif age <= 25 : cat = 'Student'
  elif age <= 35 : cat = 'Young Adult'
  elif age <= 60 : cat = 'Adult'
  else : cat = 'Elderly'

  return cat

#lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정
#get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
````
